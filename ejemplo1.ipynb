```python
X=datos["celsius"]
y=datos["fahrenheit"]
```

```python
y
```

```python
X.values.reshape(-1,1)
```

```python
X_procesada = X.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression #importar el modelo de regresion lineal
modelo=LinearRegression() #crear el modelo de regresion lineal
```

```python
#Entrenar modelo
modelo.fit(X_procesada, y_procesada)
```

```python
celsius=20
prediccion=modelo.predict([[celsius]])
print(celsius,"grados celcius son",prediccion,"grados fahrenheit")
```

```python
modelo.score(X_procesada, y_procesada)
```
